## ※ xpath

XPath는 XML 또는 HTML 문서 내에서 특정 요소나 속성을 선택하기 위해 사용되는 경로 표현 언어입니다. 웹 크롤링이나 자동화 도구에서 주로 사용되며, 요소를 효율적으로 찾을 수 있도록 도와줍니다. 일반적인 XPath는 특정 위치나 속성을 기준으로 요소를 선택하는 상대적인 경로를 사용합니다. 또한 full xpath는 루트 요소에서 시작하여 대상 요소까지의 절대적인 경로를 나타냅니다. 따라서 문서 구조가 변경되면 경로가 깨질 가능성이 높습니다.

```python
# from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By
```

```python
# 베스트 댓글
driver = webdriver.Chrome()
driver.get('https://comic.naver.com/webtoon/detail?titleId=777767&no=214&week=fri')
# soup = BeautifulSoup(driver.page_source, 'html.parser') Xpath 쓰니까 필요옶어짐

# /html/body/div[1]/div[5]/div/div/div[5]/div[1]/div[3]/div/section/ul/li[1]
# /html/body/div[1]/div[5]/div/div/div[5]/div[1]/div[3]/div/section/ul/li[2]
# /html/body/div[1]/div[5]/div/div/div[5]/div[1]/div[3]/div/section/ul/li[3]
xpath = '/html/body/div[1]/div[5]/div/div/div[5]/div[1]/div[3]/div/section/ul/li'
best_comment_elements = driver.find_elements(By.XPATH, xpath)
# best_comment_elements

for li in best_comment_elements:
    try:
        # /html/body/div[1]/div[5]/div/div/div[5]/div[1]/div[3]/div/section/ul/li[1]/div[1]/div[2]/div/p
        # /html/body/div[1]/div[5]/div/div/div[5]/div[1]/div[3]/div/section/ul/li[2]/div[1]/div[2]/div/p
        # 둘의 뿌리가 같은 부분은 아래와 같다
        # /div[1]/div[2]/div/p xpath를 받아서 일로 이렇게 들어간다! 거기에 ./로 붙이는 개념
        # /div[1]/div[2]/div/p
        comment_p = li.find_element(By.XPATH, './div[1]/div[2]/div/p')
        comment_text = comment_p.text.strip()
        print(comment_text)
        print('-'*30)
    except Exception as e:
        print(e)
```

# **3. 픽사베이 이미지 다운로드**

```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from urllib.request import Request, urlopen
```

```python
driver = webdriver.Chrome()
url = 'https://pixabay.com/ko/images/search/snow/'

# 화면 크기가 다르면 반응형때문에 XPATH가 다르게 나옴
driver.maximize_window()
driver.get(url)
time.sleep(2)
```

```python
image_xpath = '/html/body/div[1]/div[1]/div/div[2]/div[3]/div/div/div[1]/div[1]/div/a/img'
wait = WebDriverWait(driver, 10)  # 최대 10초까지 기다림. 찾으면 그 전에 넘어감
image_element = wait.until(EC.presence_of_element_located((By.XPATH, image_xpath)))
```

```python
image_url = image_element.get_attribute('src')
# 헤더는 아무 사이트 가서 가져오면 된다
image_byte = Request(image_url, headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/142.0.0.0 Safari/537.36'})
image_byte  # 객체로 가져오면 성공    
```

```python
with open('snow.jpg', 'wb') as f:
    f.write(urlopen(image_byte).read())
print('파일 저장 완료!')

driver.quit()
```

# **야놀자 리뷰 크롤링**
- [사이트](https://nol.yanolja.com/reviews/domestic/3001062)

```python
def crawl_yanolja_reviews(name, url, max_scroll=3):
    review_list = []
    driver = webdriver.Chrome()
    driver.get(url)
    time.sleep(3)  # 스크롤 기다리기

    last_height = driver.execute_script('return document.body.scrollHeight')
    # print(last_height)
    same_count = 0  # 더이상 스크롤 없는지 확인용
    scroll_times = 0

    while True:
        scroll_times += 1
        driver.execute_script('window.scrollTo(0, document.body.scrollHeight)')
        time.sleep(2) # 특정 태그를 기다리지 않기 때문에 time으로
        new_height = driver.execute_script('return document.body.scrollHeight')

        if new_height == last_height:
            same_count += 1
        else:
            same_count = 0
            last_height = new_height

        if same_count >= 3:
            print('더 이상 새로운 리뷰가 없습니다')
            break

        if scroll_times >= max_scroll:
            print('최대 스크롤 횟수 도달. 종료합니다')
            break

        # /html/body/div[1]/section/div/div[5]/div[35]/div[2]/div/div[3]/div[1]/p
        # /html/body/div[1]/section/div/div[5]/div[37]/div[2]/div/div[3]/div[1]/p

    containers = driver.find_elements(By.XPATH, '/html/body/div[1]/section/div/div[5]/div')
    # print('총 리뷰 개수:', len(containers))

    for c in containers:
        try:
            date_text = c.find_element(By.XPATH, "./div[2]/div/div[1]/div[1]/div[2]/p").text
            comment_text = c.find_element(By.XPATH, "./div[2]/div/div[3]/div[1]/p").text
        except:
            date_text=""
            comment_text=""
        review_list.append({
            "review": comment_text,
            "date": date_text
        })
    driver.quit()
    return review_list
```

```python
reviews = crawl_yanolja_reviews('서귀포','https://nol.yanolja.com/reviews/domestic/3001062', 50)
```

```python
import pandas as pd
import openpyxl

if __name__ == '__main__':
    name = '서귀포 리조트'
    url = 'https://nol.yanolja.com/reviews/domestic/3001062'
    reviews = crawl_yanolja_reviews(name, url, 100)
    df = pd.DataFrame(reviews)
    file_name = f'{name}_reviews.xlsx'
    df.to_excel(file_name, index=False)
    print('엑셀 저장 완료!')
```
